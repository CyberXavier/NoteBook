求解艾里方程时：

![](https://www.zhihu.com/equation?tex=y%27%27%3Dxy%5C%5C)

可得两个解：

![](https://www.zhihu.com/equation?tex=Ai%28x%29%3D%5Cfrac1%5Cpi%5Cint_%7B0%7D%5E%7B%2B%5Cinfty%7D%5Ccos%5Cleft%28%5Cfrac%7Bt%5E3%7D%7B3%7D%2Bxt%5Cright%29%5Ctext%7Bd%7Dt%5C%5C+Bi%28x%29%3D%5Cfrac1%5Cpi%5Cint_%7B0%7D%5E%7B%2B%5Cinfty%7D%5Cleft%5Be%5E%7B-%5Cfrac%7Bt%5E3%7D%7B3%7D%2Bxt%7D%2B%5Csin%28%5Cfrac%7Bt%5E3%7D%7B3%7D%2Bxt%29%5Cright%5D%5Ctext%7Bd%7Dt%5C%5C)

Ai(x)称为（第一）艾里函数，Bi(x)称为第二艾里函数。

```python
from scipy import special
import numpy as np
import matplotlib.pyplot as plt
```

```python
print(special.airy(1))
```

其中分别是Ai(1)，Aip(1)，Bi(1)，Bip(1)的值，Aip、Bip为Ai、Bi的导数。

```python
x = np.linspace(0,1,10)
#print(x)
print(special.airy(x))
```

其中每个数组的值与Ai，Aip，Bi，Bip相对应

```python
x = np.linspace(-10,5,1000)
plt.plot(x,special.airy(x)[0],linestyle='solid')
plt.plot(x,special.airy(x)[2],linestyle='dashed')
plt.legend(["Ai(x)","Bi(x)"],loc='best',edgecolor='blue')
plt.grid()
plt.ylim(-0.5,1)
plt.show()
```
